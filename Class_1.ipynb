```python
# == checks whether the values of two objects are equal.
# is checks whether two variables refer to the same memory object.
# So == compares values, while is compares identity.
```

```python
x = [1,2,3]
y = [1,2,3]

print(x == y)   # True (same values)
print(x is y)   # False (different memory)
```

# Data Type 

```python
# A data type defines the type of value a variable can hold and what operations can be performed on it.
# In Python, everything is an object, so every value has a data type.
```

```python
# Mutable

# A mutable object is one whose value can be changed after creation.
# Memory location → stays same
# Data → changes inside same object

# ex : list, dictionary, set
```

```python
# Immutable

# An immutable object cannot be changed after creation.
# If you try to change it → Python creates a new object in memory.

# ex : int, float , bool, string, tuple
```
