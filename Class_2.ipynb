```python
# IF - Else - we know actions from both cases (true or false)
```

```python
# Take a number input from user, if number is greter then 10 , add 1 to it else subtract 1 from it
```

```python
num = int(input("please enter number"))
print(f"number is {num}")

if num>10:
    print(f"true {num+1}")
else:
    print(f"false {num-1}")

print("exit")
```

```python
# Take a number input from user, if number is greter then 10 , add 1 to it 
# if number is smaller then 10 subtract 1 from it else print the number as it is
```

```python
num = int(input("please enter number"))
print(num)

if num>10:
    print(f"greater {num+1}")
elif num<10:
    print(f"smaller {num-1}")
else:
    print(num)
```

```python
# Develop a program that takes the temperature (in Celsius) 
# as input and classifies it based on the following criteria: 
# temperatures below 0°C should be labeled as "Extreme Cold,
# " temperatures between 0°C and 20°C as "Cold," temperatures 
# between 20°C and 30°C as "Moderate," and temperatures above 
# 30°C as "Hot." If the temperature exceeds 45°C, the program 
# should display an additional warning, "Heatwave Alert!"
```

```python
temp = float(input("Please enter the temparture (in celsius)"))
print("Temperature is ", temp)

if temp<=0:
    print("Extreme cold")
elif temp>0 and temp <= 20:
    print("cold")
elif temp>20 and temp <=30:
    print("Moderate")
elif temp>30 and temp <=45:
    print("Hot")
else:
    print("heatwave alert!")
```
